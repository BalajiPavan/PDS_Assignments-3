```python
a) Convert the text corpus into tokens.
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/Users/balajipavan/Desktop/PDS_Assignment-3/raw_data/Corona_NLP_test.csv')
data['tokens'] = data['OriginalTweet'].apply(word_tokenize)
print(data['tokens'])
```

```python
data_list=data['OriginalTweet'].values.tolist()
```

```python
data_list
```

```python
tokens = []
for tweet in data_list:
    tokens += word_tokenize(tweet)
print(tokens)
```

```python
b) Perform stop word removal.
```

```python
from nltk.corpus import stopwords

nltk.download('stopwords')

st_words = set(stopwords.words('english'))

#filter
fil_tokens = [token for token in tokens if not token.lower() in st_words]
print(fil_tokens)
```

```python
c) Count Word frequencies
```

```python
from collections import Counter

# Concat the tweets into a single string
tx = " ".join(data["OriginalTweet"].values)

# Split the string into words
wds = tx.split()

# Count the words using Counter
wd_freq = Counter(wds)

# Printing top 20 most frequent words
for word, freq in wd_freq.most_common(20):
    print(word, freq)
```

```python
d) Create word clouds.
```

```python
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
text = " ".join(tweet for tweet in data['OriginalTweet'])
wordcloud = WordCloud(width=800, height=500, random_state=21, max_font_size=110).generate(text)

plt.figure(figsize=(9, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```
